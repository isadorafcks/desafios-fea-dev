..Em criptografia, uma forma simples de codificar uma mensagem é substituir cada letra no texto por uma letra em um número fixo de posições abaixo do alfabeto. Isto é um tipo de cifra.

Por exemplo, com um deslocamento (shift) de 3, **'A'** seria substituído por **'D'**, **'B'** se tornaria **'E'** e, assim por diante... No final do alfabeto mantendo o mesmo exemplo **'Z'** se tornaria **'C'**
